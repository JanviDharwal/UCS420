```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
"""
Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
-> Generate x values from -10 to 10 using np.linspace().
->Compute y values for:
o y= M⋅x^2
2
o y=M⋅sin(x)
-> Plot both functions in a single figure:
-> Use different colors and line styles.
-> Add legend, grid, and title.
"""
M=int(input("enter val"))
#geneating x values
x=np.linspace(-10,10,100)#start,stop number
#computing y values
y=M*x**2
y1=M*np.sin(x)
plt.plot(x,y,color='blue',label='y=M*x^2',linestyle='--',linewidth=2)
plt.plot(x,y1,color='red',label='y=M*sin(x)',linestyle='--',linewidth=4)
plt.legend()
plt.grid(True)#enables grid on the plot
plt.title('Mathimatical FUN')
plt.show()
```

```python
'''
Q.2 Create a dataset of five subjects and scores.
->Convert it into a Pandas DataFrame.
->Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.
'''
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math', 'English', 'Physics', 'Chemistry', 'Biology'],
    'Score': [85, 92, 78, 88, 95]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.barplot(x='Subject', y='Score', data=df, palette='Set2')

for index, value in enumerate(df['Score']):
    plt.text(index, value + 1, str(value), ha='center', va='bottom', fontsize=12)


plt.title('Scores in Different Subjects', fontsize=16)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
'''
Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
->Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
->Customize with titles, axis labels, and grids.
'''
import numpy as np
import matplotlib.pyplot as plt

roll_number = 12345
np.random.seed(roll_number)
data = np.random.randn(50)
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].plot(np.cumsum(data), color='blue', label='Cumulative Sum')
axs[0].set_title('Cumulative Sum Plot')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Cumulative Sum')
axs[0].grid(True)
axs[0].legend()
noise = np.random.randn(50)
axs[1].scatter(range(50), data + noise, color='red', label='Scatter with Noise')
axs[1].set_title('Scatter Plot with Random Noise')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Values with Noise')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)

# Check column names
print(df.columns)

# Ensure 'Month' is treated properly
df.rename(columns={'month_number': 'Month'}, inplace=True)

# Convert numeric month to actual names
month_mapping = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May',
                 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October',
                 11: 'November', 12: 'December'}
df['Month'] = df['Month'].map(month_mapping)

# Task 1: Line plot for Total Profit of all months
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='total_profit', data=df, marker='o', color='blue')
plt.title('Total Profit for All Months', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Profit', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# Task 2: Multiline plot for all product sales data
product_columns = df.columns[1:-1]  # Excluding 'Month' and 'total_profit'
plt.figure(figsize=(12, 6))
for product in product_columns:
    sns.lineplot(x='Month', y=product, data=df, label=product)

plt.title('Product Sales Over the Months', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.legend(title='Products', loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Task 3: Bar chart for all features
plt.figure(figsize=(12, 6))
df.drop(columns=['Month']).sum().plot(kind='bar', color='teal')
plt.title('Bar Chart for All Features/Attributes', fontsize=14)
plt.xlabel('Features', fontsize=12)
plt.ylabel('Sum of Values', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
