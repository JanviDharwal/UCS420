```python
import numpy as np

import numpy as np
import matplotlib.pyplot as plt

# Generate x val
x = np.linspace(-10, 10, 100)

# Compute y val
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)


plt.figure(figsize=(12, 8))

#a Plot x^2
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r', label=r'$y = x^2$')
plt.title("Square Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

#b Plot sin(x)
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g', label=r'$y = \sin(x)$')
plt.title("Sine Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

#c Plot e^x
plt.subplot(2, 2, 3)
plt.plot(x, y3, 'b', label=r'$y = e^x$')
plt.title("Exponential Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

#d Plot log(|x| + 1)
plt.subplot(2, 2, 4)
plt.plot(x, y4, 'm', label=r'$y = \log(|x| + 1)$')
plt.title("Logarithmic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

#e Adjust layout and show plot
plt.tight_layout()
plt.show()
```
